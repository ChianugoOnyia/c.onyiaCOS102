```python
import tkinter as tk
from tkinter import messagebox

class Delivery:
    def totalcost(self, location, weight):
        location = location.upper()

        if location == "PAU":
            if weight >= 10:
                return 2000
            else:
                return 1500

        elif location == "EPE":
            if weight >= 10:
                return 5000
            else:
                return 4000

        else:
            return None

system = Delivery()

def submit():
    location = entry_location.get()
    weight = entry_weight.get()

    if weight == "":
        messagebox.showerror("Error", "Enter package weight")
        return

    weight = float(weight)

    cost = system.totalcost(location, weight)

    if cost is None:
        messagebox.showerror("Error", "Invalid location")
    else:
        messagebox.showinfo(
            "Delivery Cost",
            f"Location: {location}\nWeight: {weight}kg\nAmount to Pay: N{cost}"
        )


root = tk.Tk()
root.title("Delivery System")
root.geometry("400x250")  


tk.Label(root, text="Enter Location: (PAU / EPE)").pack(pady=5)
entry_location = tk.Entry(root)
entry_location.pack()

tk.Label(root, text="Enter Package Weight (kg):").pack(pady=5)
entry_weight = tk.Entry(root)
entry_weight.pack()

# Button
tk.Button(root, text="Calculate Cost", command=submit).pack(pady=15)

root.mainloop()
```
