```python

P = 10000      
R = 5          
T = 3          

simple = P * (1 + (R / 100) * T)
print("Simple Interest Amount =", simple)

P = 10000
R = 5
T = 3
n = 4         

compound = P * (1 + (R / (100 * n))) ** (n * T)
print("Compound Interest Amount =", compound)

PMT = 2000     
R = 5
T = 3
n = 4          

r = R / (100 * n)
annuity = PMT * (((1 + r) ** (n * T) - 1) / r)
print("Annuity Amount =", annuity)
```
