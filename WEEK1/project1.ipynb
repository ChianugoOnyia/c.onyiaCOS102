```python


print("PROJECT1")

print("Choose an option:")
print("1: Simple Interest")
print("2: Compound Interest")
print("3:Annuity")

choice = input("What u wanna do? ")

if choice == "1":
    P = float(input("Principal : "))
    R = float(input("Rate (%): "))
    T = float(input("Time (years): "))
    
    simple = P * (1 + (R / 100) * T)
    print("Simple Interest Amount =", round(simple, 2))

elif choice == "2":
    P = float(input("Principal : "))
    R = float(input("Rate of interest(%): "))
    T = float(input("Time (years): "))
    
    compound = P * (1 + (R / 100)) ** T
    print("Compound Interest Amount =", round(compound, 2))

elif choice == "3":
    PMT = float(input("(PMT): "))
    R = float(input(" Rate (%): "))
    T = float(input("Time (years): "))
    
    annuity = PMT * ((((1 + R/100) ** T) - 1) / (R/100))
    print("Annuity Amount =", round(annuity, 2))


else:
    print("Invalid choice. Please run the program again")
    

    
```
