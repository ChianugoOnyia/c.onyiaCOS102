```python
import tkinter as tk
from tkinter import messagebox as msgbox
from PIL import Image, ImageTk
import csv

def welcomemessage(Name, Department, members):
   
    window = tk.Toplevel(root)
    window.title("Admin Box")
    window.geometry("500x300")

  
    label_1 = tk.Label(window, text=f"Welcome {Name}\n")
    label_1.pack()

    label_2 = tk.Label(window, text="GIG LOGISTICS")
    label_2.pack()


    if members:
        other_members = "\n".join(members)
        label_3 = tk.Label(window, text=f"Other members in {Department}:\n{other_members}")
        label_3.pack()
    else:
        label_3 = tk.Label(window, text="You are the only member in your department.")
        label_3.pack()


def submit():
    Name = Name_entry.get()
    Department = Department_entry.get()

    found = False
    department_members = []

    with open("GIG-logistics(in).csv", "r") as file:
        reader = csv.reader(file)
        headers = next(reader)  

        for row in reader:
          
            if Name.strip() == row[1].strip() and Department.strip() == row[2].strip():
                found = True
            
            if Department.strip() == row[2].strip() and Name.strip() != row[1].strip():
                department_members.append(row[1].strip())

    if found:
        welcomemessage(Name, Department, department_members)
    else:
        msgbox.showerror("Login", "Invalid name or department")



root = tk.Tk()
root.title("Login Form")
root.geometry("500x200")

Name_label = tk.Label(root, text="Name:")
Name_label.pack()
Name_entry = tk.Entry(root)
Name_entry.pack()

Department_label = tk.Label(root, text="Department:")
Department_label.pack()
Department_entry = tk.Entry(root)
Department_entry.pack()

submit_button = tk.Button(root, text="Submit", command=submit)
submit_button.pack()

root.mainloop()
```
