```python
import tkinter as tk
from tkinter import messagebox



class Zenith:
    def mutualservices(self):
        return ["Lines of credit", "Investment management and accounts", "Insurance"]

    def uniqueservices(self):
        return []


class Retailbanking(Zenith):
    def uniqueservices(self):
        return [
            "Retirement and education accounts","Loans and mortgages","Checking and saving" ]


class Globalbanking(Zenith):
    def uniqueservices(self):
        return [
            "Multi-currency management","Foreign currency accounts","Foreign currency credit cards","Transborder advisory services","Liquidity management" ]


class Commercialbanking(Zenith):
    def uniqueservices(self):
        return ["Advisory services"]


# Function
def showservices():
    retailnames = {"mary"}
    globalnames = {"agatha"}
    commercialnames = {"noel"}
    name = name_entry.get().strip().lower()
    division = division_var.get()

    # Create object based on division
    if name in retailnames and division == "Retailbanking":
        obj = Retailbanking()
        services = obj.mutualservices() + obj.uniqueservices()
    elif name in globalnames and division == "Globalbanking":
        obj = Globalbanking()
        services = obj.uniqueservices()
    elif name in commercialnames and division == "Commercialbanking":
        obj = Commercialbanking()
        services = obj.mutualservices() + obj.uniqueservices()
    else:
        messagebox.showerror("Error", "Wrong name or division")
        return

    # Get services AFTER object is created
    

    # Build result text
    result = f"Employee: {name}\nDivision: {division}\n\nServices:\n"
    result += "\n".join(f"- {s}" for s in services)

    # Show output
    messagebox.showinfo("Services", result)
    output_label.config(text=result)


# GUI
root = tk.Tk()
root.title("Zenith Bank Services")
root.geometry("400x400")

tk.Label(root, text="Employee Name:").pack()
name_entry = tk.Entry(root)
name_entry.pack()

tk.Label(root, text="Select Division:").pack()
division_var = tk.StringVar()

tk.Radiobutton(root, text="Retail", variable=division_var, value="Retailbanking").pack()
tk.Radiobutton(root, text="Global", variable=division_var, value="Globalbanking").pack()
tk.Radiobutton(root, text="Commercial", variable=division_var, value="Commercialbanking").pack()

tk.Button(root, text="Submit", command=showservices).pack(pady=10)

output_label = tk.Label(root, text="", justify="center")
output_label.pack()

root.mainloop()
```
